```python
def att_code_type(df_attendance):
    if (df_attendance['Abs/Att Type']) ==   'Z010':
        return 'At Work'
    elif (df_attendance['Abs/Att Type']) == 'Z055':
        return 'Sick Leave'
    elif (df_attendance['Abs/Att Type']) == 'Z070':
        return 'Public Holiday'
    elif (df_attendance['Abs/Att Type']) == 'Z045':
        return 'Vacation'
    elif (df_attendance['Abs/Att Type']) == 'Z020':
        return 'Extra Hours'
    elif (df_attendance['Abs/Att Type']) == 'Z051':
        return 'Comp Off'
    elif (df_attendance['Abs/Att Type']) == 'Z074':
        return 'Optional Holiday'     
    elif (df_attendance['Abs/Att Type']) == 'Z054':
        return 'Maternity Leave'
```

## 5. myTe Submission on time or not?

```python
pd.to_datetime(dataframe['Processed Date in SAP (in CST)']).head(12)
```

```python
pd.to_datetime(dataframe['Submitted Date in SAP (in CST)']).head(12)
```

```python
time_diff = (pd.to_datetime(dataframe['Processed Date in SAP (in CST)'])-pd.to_datetime(dataframe['Submitted Date in SAP (in CST)']))
```

```python
# This inserts a new column that shows how early myTE was submitted by an Employee.
df_attendance.insert(16,'myTE_Submission_Before_X_Days',pd.DataFrame(time_diff.dt.days),True) 
```

```python
df_report_submission = df_attendance.groupby(['TR Period','Submitted Period','myTE_Submission_Before_X_Days']).size().reset_index().rename(columns = {'TR Period': 'Time Report Period', 0:'Fortnight Working Days'})
```

```python
df_report_submission1 = df_report_submission[(df_report_submission['Time Report Period'] == df_report_submission['Submitted Period'])]
```

```python
df_report_submission1
```

```python
df_report_submission1['Submitted Period'].dt.date
```

```python
ax = sns.barplot(x=df_report_submission1['Submitted Period'].dt.date,y="myTE_Submission_Before_X_Days", data=df_report_submission1, hue = 'Fortnight Working Days')
ax.set_xticklabels(ax.get_xticklabels(), rotation=80)
```

### Prior Period Adjustments highlight

```python
df11 = df_report_submission[(df_report_submission['Time Report Period'] != df_report_submission['Submitted Period'])][['Time Report Period','Submitted Period']]
```

```python
df11
```

```python
ax = sns.countplot(x=df11['Submitted Period'].dt.date,hue = df11['Time Report Period'].dt.date, data=df11,palette = 'Set2')
```

Insights from graph:

1. This graph Says, Roger has done PPA for January Month (1st and 2nd Fortnight ) in 1st fortnight of May 19(Submitted Period).
2. Similarly, PPA for 2nd fortnight in January was done again in 1st fortnight of Feb19.
3. Also PPA for 1st fortnight in Feb was done in 2nd Fortnight in Feb.
